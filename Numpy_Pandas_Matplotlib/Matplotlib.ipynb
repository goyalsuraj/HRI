**Sine and Cosine Plot**

```
# Define the time interval of the sine wave

time = np.linspace(0,10,50)
#time = np.arange(0,2*np.pi,0.1)
 

# Define amplitude of the sine wave and cosine wave

amplitude1   = np.sin(time)
amplitude2 = np.cos(time)
```

```
# Plot the sine and cosine wabve
plt.plot(time, amplitude1, label="sine")
plt.plot(time, amplitude2, '^', label="cosine")

# Labelling the plot
plt.title('Sine and Cosine wave')

#Legend
plt.legend(loc="upper right")
# Labelling the Axis
plt.xlabel('Time')
plt.ylabel('Amplitude')
#plt.ylim(0,1)
```

**Multiple Histograms Plot**

```
x1 = np.random.normal(0, 0.8, 1000)
x2 = np.random.normal(-2, 1, 1000)
x3 = np.random.normal(3, 2, 1000)

args = dict(histtype='stepfilled', alpha=0.3, normed=True, bins=40)

plt.hist(x1, **args)
plt.hist(x2, **args)
plt.hist(x3, **args)
```
