#Creating Lists
-Lists can be modified but have to follow a specific order

-Tuples are similar to lists and can be created with round brackets but cannot be manipulated, i.e. they are unchangeable

-They can contain dataTypes like integers, strings etc. in a single list

-Elements in a list are indexed based on the defined sequence. Indexing of a list starts with 0 being the first index with each element assigned a definite distinct location.

```
#Defining a list
sports_list = ["soccer", "football", "foosball", "basketball"]
print'List of sports\n',sports_list
```

```
print'3rd item on the list is:',sports_list[2]
```

#Replacing item in the list

```
#Replacing foosball with a different sport
sports_list[2]='baseball' 
print'New list of sports\n',sports_list
```

```
for i in sports_list:
  print(i)
```

```
#Identifying the index or location of item in the list
sports_list.index('soccer')
```

```
#Identifying last item in a list
print'Last item in the list:',sports_list[-1]
```

```
#Printing 1st two items in the list
print'1st two sports in the list:',sports_list[0:2] 

#Remember in a range [0:x] will print till x-1 as last item is not counted
```

**Conditional if statement to return a message based on condition being true or not**

```
if 'cricket' not in sports_list:
  print('Cricket is not in the list')
else:
  print('Found it!')
```

```
#Length function helps to identify number of items stored in the list
print'Number of items in the sports list:',len(sports_list)
```

**Append method adds an item to end of the list**

```
sports_list.append('cricket')

#Check to see item got added to the list
if 'cricket' not in sports_list:
  print('Cricket is not in the list')
else:
  print('Found it!')
```

**Remove items from the list**

```
sports_list.remove('cricket')

if 'cricket' not in sports_list:
  print('Cricket is not in the list')
else:
  print('Found it!')
```

```
#Another method of appending a list using a '+' operator
sports_list=sports_list+['dodgeball']
print'New sport added at the end of the list',sports_list
```

```
#Multiplying to regenerate items in the list
sports_list*=2
print 'Sports list after doubling:\n',sports_list
```

#Tuple is ordered like a list but is unchangeable. 
Note: Tuples use parentheses while Lists use square brackets

```
#Defining a tuple with round brackets
sports_list = ("soccer", "football", "foosball", "basketball")
print(sports_list)
```

**Tuple cannot be modified like a list**

```
sports_list[2]='baseball' 
print(sports_list)
```
