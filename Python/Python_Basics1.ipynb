#Number Types: int, float and complex

1.   Integer is a whole number, positive or negative, without decimals, of unlimited length

2.   Floating point number can be positive or negative, containing one or more decimals

3.   Complex numbers have a "j" as the imaginary part






```
a = 10    # int
b = 3.14159265359  # float
c = 10j   # complex
```

```
a,b,c=10, 3.14159265359, 10j

print type(a), type(b), type(c)
print(a+b+c)
```

#String Manipulation

```
a = "It's a wonderful day outside."
print'Statement defined:',a

#print only the first string character
print'\nFirst character in the statement:',a[0]
```

```
print'First 4 characters in the string:',a[0:4]

print'Another way of printing the entire string:',a[0:]
```

```
#Number of characters in the string
print'Number of characters:', len(a)
```

```
print'Printing in all uppercase:\n',a.upper()
""" Converting text to upper case
    and lower case letters """
print'\nPrinting in all lowercase:\n',a.lower()
```

**format()** takes the arguments, changes the types and places them in the string where the placeholder {} is

```
age=80
experience=5
text="John is old, since he is {}, experience {} "
print text.format(age, experience)
```

**format()** multiple arguments

```
age=80
car=2
price= 5
text="John is old, since he is {} but his car is only {} years old that he bought for ${} "
print text.format(age,car, price)
```
