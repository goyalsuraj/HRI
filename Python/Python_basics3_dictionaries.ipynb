**Dictionary  are defined with curly brackets, consisting of keys and values.
This is the primary difference with lists in which elements are accessed by their position in the list, via indexing.**

```
bike_dict =	{
  "brand": "Pinarello",
  "model": "Dogma",
  "year": 1999
}
print 'Dictionary with attributes for a bike\n',bike_dict
```

**Indexing dictionary with the key to get the value associated with it.**

```
x = bike_dict["model"]
print 'Value for the key model is:',x
```

**Changing Bike's Model Year or the value**

```
bike_dict =	{
  "brand": "Pinarello",
  "model": "Dogma",
  "year": 1999
}
bike_dict["year"] = 2019
```

```
x = bike_dict["year"]
print 'Value for the key year is:',x
```

**Displaying all the Keys Only **

```
#Printing all the keys in the dictionary
for x in bike_dict:
  print x
```

**Displaying Keys and Values together**

```
#Printing all keys and values
for x in bike_dict:
  print x,'-', bike_dict[x]
```

**items() function to  return a list with all dictionary keys with values**

```
print bike_dict.items()
```

**items() to return a list 1 key at a time**

```
#Printing 1 key at a time
for x, y in bike_dict.items():
  print(x,y)
```

**Like lists key is checked in the dictionary with a conditional if statement**

```
if "model" in bike_dict:
  print("Yes, 'model' is a dictionary key.")
```

**len() gives the total length of the dictionary.**

```
print(len(bike_dict))
```

**Adding attribute or key and value**

```
bike_dict =	{
  "brand": "Pinarello",
  "model": "Dogma",
  "year": 1999
}
bike_dict["color"] = "red"
print(bike_dict)
```

**Removing a specific entry**

```
bike_dict =	{
  "brand": "Pinarello",
  "model": "Dogma",
  "year": 1999
}
bike_dict.pop("model")
print(bike_dict)
```

**Creating another dictionary Copy**

```
bike_dict =	{
 "brand": "Pinarello",
  "model": "Dogma",
  "year": 1999
}
newdict = bike_dict.copy()
newdict['material']='carbon fiber'
print(newdict)
```
