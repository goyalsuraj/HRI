**Create a function to calculate compound interest which takes in input arguments for principal, rate of interest, time period and compounding frequency**

CI=P(1+(R/100*N))^(NT)

```
import math

#Compound Interest Function
def CI(principal, interest, time_period, compound_freq):
  ci_expected = principal * (math.pow((1 + interest / (100*compound_freq)), time_period*compound_freq))
  print("Expected Compound Interest for Principal Amount ${0} with {1} % interest over {2} year(s) is ${3}".format(principal, interest, time_period, ci_expected))
  
CI(1000,10,1,12)
```

**Request use to enter a positive integer and check to see if it is a prime number or not**

```
prime_num=input("Enter a positive integer: ")
for i in range(2,prime_num):
  if prime_num%i==0:
    
    print("{0} is not a prime number").format(prime_num)
    factor=prime_num/i
    print("{0}*{1} = {2}".format(i,factor,prime_num))
    break
    
  else:
    print("{0} is a prime number").format(prime_num)
    
    break
```

**Program to take user input as a list and to calculate mean and standard deviation based on the data provided**


```
def user_data():
  # creating an empty list 
  user_input = [] 
  # number of elements as input 
  num = int(input("Enter number of elements : ")) 
  
  for i in range(0, num): 
    items = input()
    #Adding items to the empty list  
    user_input.append(items)  
  return user_input
```

```
#Adding the math library
import math
def avg(data):
  """Calculating the average based on the user input"""
  n=float(len(data))
  return (sum(data)/n)

avg(user_input)

def sd(data):
  """ Standard deviation function"""
 
  n=len(data)
  sd_cal=0.0
  #Using the avg function
  mean=avg(data)
  
  
  for i in data:
    sd_cal+=math.pow((float(i)-mean),2)
    #sd_cal+=((float(i)-mean)**2)
    
  sd_cal = math.sqrt(sd_cal / float(n-1))
  print("For User list: {0} \nStandard Deviation is: {1} and Mean is: {2}").format(data, sd_cal, mean)

sd(user_data())
```
