#Loops in Python
**Repetitive execution of the same block of code over and over is referred to as an iteration.
There are two types of iteration:**

1. Definite iteration: Number of repetitions are specified explicitly in advance.

2. Indefinite iteration: Code block executes until some condition is met.


```
#While loop runs between the initial value and runs till the end condition
i = 1
#Loop Range 1-19
while i < 20:
  #printing first and then incrementing the value
  print(i)
  i += 2
  
print'\nLoop completed!' 
```

**Changing the order of print statement with incrementing**

```
i = 1

while i < 20:
  #incrementing the value and then printing
  i += 2
  print(i)
  
print'\nLoop completed!' 
```

**Break Statement can stop the loop even if the while condition is true**

```
i = 1
while i < 20:
  print(i)
  if i== 12:
    break
  i+=1
print'\nLoop completed!' 
```

**Continue Statement can stop the current iteration, and continue with the next**

```
i = 1
while i < 20:
  i+=1
  # Continue to the next iteration if i is 12 or 13
  if i== 12 or i==13:
    continue
  print(i)
  
print'\nLoop completed!'  
```

```
for val in "rhetorical":
    if val == "i":
        break
    print(val)

print("\nEnd of the program!")
```

```
for val in "rhetorical":
    if val == "i":
        continue
    print(val)

print("\nEnd of the program!")
```

**Using while to print a statement until the counter is valid**

```
counter = 0

while counter < 3:
    print("While loop is running.")
    counter = counter + 1
else:
    print("\nLoop over!")
```

**Revising lists**
Adding for loop which has a range defined vs while loop which is based on an end condition

```
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
for j in months:
  print(j)
```

```
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sum=0
for j in months:
  # Exit the loop if below condition is met
  if j=='Jun':
    break  
  sum=sum+1
  print sum, j
  
```

```
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
sum=0
for j in months:
   # Continue the loop if below condition is met and skip the value
  if j=='Jun':
    continue
  sum=sum+1
  print sum, j
```
