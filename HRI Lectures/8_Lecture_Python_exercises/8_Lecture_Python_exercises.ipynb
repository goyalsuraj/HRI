**3. Based on the user input, print numbers 1 to the number defined by the user (range) and identify whether each of them is even or odd**

Output

```
num=int(input("Please provide a positive integer number "))

for i in range(1,num+1):
  if i%2==0:
    print"Even",i
  else:
    print"Odd ",i
```

**4. Write a program to compute the factorial of a number based on the user input.**

```
fact_num=input("Please provide a positive integer ")
fact_sum=1
for i in range(1,fact_num+1):
  fact_sum=fact_sum*i
print(fact_sum)
  
```

**5. CI=P*((1+(Rate_of_interest/(100*N))^(N*T)**

User inputs: Principal, Rate, Compounded Frequency, T is the time

Calculate Compound Interest (CI)

```
def compound_interest(principle, rate, cf, time): 
  
    # Calculates compound interest  
    CI = principle * (pow((1 + rate /(cf*100)), (cf*time))) 
    print "Compound interest is", CI
  
# Function Call 
compound_interest(10000, 10.25,4, 5) 
```

**6. Print a list of prime numbers from 1 to 100 inclusive**

```
primes = []
for prim_num in range(2, 100):
    
    # Assume number is prime until shown it is not. 
    isPrime = True
    for num in range(2, prim_num):
        if prim_num % num == 0:
            isPrime = False
            break
      
    if isPrime:
        primes.append(prim_num)
print(primes)
```
